```python
# Question 3
# Method 1: Display data types using .dtypes attribute
print("Data types of each column: ")
print(df.dtypes)

# Method 2: Display data types in a more readable format
print("\nData Types (Detailed View):")
for column in df.columns:
    print(f"{column}: {df[column].dtype}")

# Method 3: Count columns by data type
print("\nCount of columns by data type:")
print(df.dtypes.value_counts())

# Method 4: Display data type information using info()
print("\nDetailed Data Type Information (info):")
df.info()
```
