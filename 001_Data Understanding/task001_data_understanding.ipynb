### Analysis of Question 2:

#### هدف سوال
هدف از این سوال، شناسایی کامل تمام ویژگی‌های موجود در دیتاست است. دانستن نام ستون‌ها به ما کمک می‌کند تا:

* درک کنیم چه نوع اطلاعاتی در دیتاست جمع‌آوری شده است.

* ویژگی‌های عددی و دسته‌ای را تشخیص دهیم.

* برای مراحل بعدی (تحلیل، مصورسازی و مدل‌سازی) برنامه‌ریزی کنیم.

* از دیدگاه پزشکی و کسب‌وکار، معنای هر ویژگی را درک کنیم.

#### نکات تحقیقی:
1. تعداد و ساختار ستون‌ها: دیتاست دقیقاً دارای ۱۴ ستون است که شامل ۱۳ ویژگی ورودی و ۱ متغیر هدف (target) می‌باشد. این ساختار برای یک مسئله طبقه‌بندی نظارت شده (Supervised Classification) مناسب است.

2. طبقه‌بندی ویژگی‌ها بر اساس نوع داده:

    * ویژگی‌های عددی پیوسته (Continuous Numerical):

        - age (سن بیمار): محدوده ۲۹ تا ۷۷ سال

        - trestbps (فشار خون استراحت): بر حسب میلی‌متر جیوه

        - chol (کلسترول سرم): بر حسب mg/dl

        - thalach (حداکثر ضربان قلب): بر حسب ضربه در دقیقه

        - oldpeak (افت ST ناشی از ورزش): نسبت به حالت استراحت

    * ویژگی‌های عددی گسسته/دسته‌ای (Categorical/Discrete):

        - sex (جنسیت): ۰ = زن، ۱ = مرد

        - cp (نوع درد قفسه سینه): ۰-۳ (۴ نوع مختلف)

        - fbs (قند خون ناشتا): ۰ = ≤ ۱۲۰ mg/dl، ۱ = > ۱۲۰ mg/dl

        - restecg (نتیجه نوار قلب در حالت استراحت): ۰-۲ (۳ نوع مختلف)

        - exang (آنگین ناشی از ورزش): ۰ = خیر، ۱ = بله

        - slope (شیب ST): ۰-۲ (۳ نوع مختلف)

        - ca (تعداد رگ‌های اصلی): ۰-۴

        - thal (نوع تالاسمی): ۰-۷ (مقادیر مختلف)

    * متغیر هدف:

        - target (بیماری قلبی): ۰ = بدون بیماری، ۱ = دارای بیماری

3. نکات مهم از دیدگاه پزشکی و کسب‌وکار:

- ویژگی cp (نوع درد قفسه سینه) یکی از مهم‌ترین ویژگی‌های بالینی است که در تشخیص بیماری قلبی نقش کلیدی دارد.

- ویژگی thalach (حداکثر ضربان قلب) نشان‌دهنده عملکرد قلبی-عروقی بیمار است.

- ویژگی oldpeak یک شاخص مهم در تست ورزش است که نشان‌دهنده کاهش اکسیژن رسانی به قلب می‌باشد.

- ویژگی ca (تعداد رگ‌های اصلی) نتیجه آنژیوگرافی است که معمولاً یک ویژگی تشخیصی قوی محسوب می‌شود.

#### علت و دلیل استفاده از این کد
1. استفاده از df.columns: این attribute ساده‌ترین و استانداردترین روش برای دریافت نام ستون‌ها است. خروجی آن یک Index object از نام ستون‌ها است.

2. استفاده از df.columns.tolist(): تبدیل نام ستون‌ها به لیست، کار با آنها را در کدنویسی‌های بعدی آسان‌تر می‌کند (مانند حلقه‌ها، شرط‌ها و انتخاب ستون‌ها).

3. نمایش ستون‌ها با ایندکس: این روش به ما کمک می‌کند تا موقعیت هر ستون را بشناسیم و در صورت نیاز با ایندکس به ستون‌ها دسترسی داشته باشیم.

4. نمایش head(2): نمایش چند ردیف اول به درک بهتر محتوای هر ستون کمک می‌کند و نشان می‌دهد که هر ستون چه نوع داده‌ای را ذخیره می‌کند.

5. تعداد کل ستون‌ها: این شمارش تأیید می‌کند که تعداد ستون‌ها با آنچه در سوال ۱ مشاهده کردیم (۱۴ ستون) مطابقت دارد.
